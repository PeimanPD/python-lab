```python
vowels = ["i", "e", "a", "o", "u"]
for v in vowels:
    print(f"Vowel: {v.upper()}")
```

```python
def vowel_count(words, target="i"):
    """Count how many words contain the target vowel."""
    return sum(target in w for w in words)

sample_words = ["sit", "seat", "set", "sot", "suit"]
vowel_count(sample_words)
```

```python
speakers = ["F01","F02","M01","M02"]
vowel = ["i","e","a","o"]
F1 = [300,400,500,600]
F2 = [2300,2000,1700,1400]
```

```python
import pandas as pd
df = pd.DataFrame({"Speaker": speakers[:4],
                   "Vowel": vowel,
                   "F1": F1,
                   "F2": F2})
df
```

```python
## Step 5 – Plotting Vowel Space
#F1 and F2 values plotted with inverted F2 axis to mirror vowel charts in phonetics.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x="F1", y="F2", hue="Vowel", s=100)
plt.gca().invert_yaxis()
plt.title("Sample Vowel Space")
plt.xlabel("F1 (Hz)")
plt.ylabel("F2 (Hz)")
plt.show()
plt.savefig("D:/Speech-ai-roadmap/python-lab/figures/day1_vowelspace.png", dpi=300)
```
